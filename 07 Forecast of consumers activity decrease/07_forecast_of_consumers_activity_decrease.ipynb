Описание проекта
        
Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.
    
    Решение задачи:
        1. Нужно построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца.
        2. В исследование нужно включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца.
        
        3. Используя данные модели и данные о прибыльности клиентов, нужно выделить сегменты покупателей и разработать для них персонализированные предложения.
        
        Описание данных
Данные для работы находятся в нескольких таблицах. market_file.csv
Таблица, которая содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении.
- id — номер покупателя в корпоративной базе данных.
- Покупательская активность — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».

- Ошибка_сервиса — число сбоев, которые коснулись покупателя во время посещения сайта.
- Страниц_за_визит — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

налицо дисбаланс классов. Чтобы корректно разбить несбалансированный датасет на тренировочную и тестовую выборки, нужно использовать стратификацию данных. При формировании тестовой и тренировочной выборок важно сохранить исходное распределение целевого признака. Для этого нужно указать значения целевого признака в настройке stratify функции train_test_split().

При дисбалансе классов для оценки качества модели лучше использовать F1-меру или ROC-AUC, т.к. при оценке качества модели важно внимательно подбирать метрику - не все из них учитывают дисбаланс классов. ## 6. Использование пайплайнов

### 6.1 Подготовка данных
Во время подготовки данных используем ColumnTransformer. 
Количественные и категориальные признаки обработаем в пайплайне раздельно. Для кодирования категориальных признаков используем как минимум два кодировщика, для масштабирования количественных — как минимум два скейлера.

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25


X_train, X_test, y_train, y_test = train_test_split(
    df_main.drop(['Покупательская активность'], axis=1),
    df_main['Покупательская активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = df_main['Покупательская активность'])

X_train.shape, X_test.shape
#display(X_train, y_train)
```

```python
df_main.head()
```

```python
# создаём списки с названиями признаков
#le_columns = ['Покупательская активность']
ohe_columns = ['Разрешить сообщать', 'Популярная_категория']
ord_columns = ['Тип сервиса']
num_columns = ['Маркет_актив_6_мес', 'Маркет_актив_тек_мес', 'Длительность', 'Акционные_покупки', 'Средний_просмотр_категорий_за_визит', \
               'Неоплаченные_продукты_штук_квартал', 'Ошибка_сервиса', 'Страниц_за_визит', 'Выручка_-1_мес', \
               'Выручка_-2_мес', 'Выручка_текущий_мес', 'предыдущий_месяц_мин', 'текущий_месяц_мин']

# кодируем целевой признак
le = LabelEncoder()

y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)


# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='error', sparse=False))
    ]
    )

# создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум']                    
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough',
    #verbose_feature_names_out=True
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2,5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1',
            class_weight='balanced'
            #C=0.1,
            #solver='saga'
        )],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели SVC()
    {
        'models': [SVC(
            random_state=RANDOM_STATE, 
            kernel = 'poly', 
            degree = 2,
            C = [0.1,1,10,100],
            gamma = [0.1,1,10,100],
        )], 
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train_encoded)

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)

#прогоняем пайп на тестовых данных (кодируем целевой признак)

# проверьте работу модели на тестовой выборке
# рассчитайте прогноз на тестовых данных

y_test_pred = randomized_search.predict(X_test)
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, randomized_search.predict_proba(X_test)[:,1])}')

```

 ROC-AUC — это площадь под ROC-привой. Она оценивает качество решений модели в наиболее общем виде, учитывая все возможные пороговые значения. Поэтому она не зависит от того, какой порог выбирается, это самая общая оценка итогов работы модели. Метрику ROC_AUC можно использовать в случае, когда объектов с одним значением целевого признака больше, чем с другими.

Как можно использовать эти наблюдения при моделировании: 
можно попробовать удалить неважные признаки и обучить лучшую модель без них, таким образом, возможно, повыcится ее точность и эффективность (быстрота) работы. Мы этого делать не будем, т.к. на этапе обучения моделей уже применяется L1-регуляризация для модели линейной регрессии, которая оказалась лучшей, поэтому лишние признаки уже отсеяны, остались только те, которые в той или иной степени, но влияют на результат.

как можно использовать эти наблюдения при принятии бизнес-решений:

    1. Очевидно, сильнее всего на покупательскую активность влияет количество просмотренных страниц за визит. Учитывая, что по условиям задачи, "Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов.", акцент нужно делать не на привлечении новых клиентов, а на повышение просмотренных покупателем страниц на сайте (чем больше страниц покупатель отсмотрел, тем больше признак склоняет модель отнести объекты к классу 0, т.е. покупательская активность остается на прежнем уровне (НЕ снижается).
    
    2. Второй по важности признак - выручка в препредыдущем месяце. Это, видимо, связано с тем, что сбор информации по целевому признаку (снизилась активность или нет) осуществлялся на квартальной основе, т.е. за 3 мес. (не зря же по условиям задачи "Нужно построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца."). Из этого следует 2 вывода:
    
    а) либо поменять методику маркировки целевого признака в тренировочной выборке (с квартальной на месячную), с тем, чтобы выручка на него влияла не только 2 месячной давности, но и в предыдущем и в текущем месяцах - это позволит давать более точные прогнозы снижения активности,
    
    б) при действующей методике - выручка, очевидно, является одним из важнейших признаков, но ориентироваться при прогнозах можно на  выручку 2-хмесячной давности.
    
    В любом случае, учитывая, что выручка - один из важнейших признаков активности, важно введение корреляции "плюшек" от компании (каких именно, сейчас трудно сказать, возможно, скидок) с выручкой, т.е. стимуляция большей выручки для неснижения активности покупателей.
    
    3. Учитывая, что третьим по значимости признаком является "общее число неоплаченных товаров в корзине за последние 3 месяца", причем чем больше товаров лежит в корзине, тем больше признак склоняет модель отнести объекты к классу 1, т.е. активность СНИЖАЕТСЯ, можно сделать вывод, что, вероятно, большое число неоплаченных товаров рассеивает внимание покупателя и не дает сконцентрироваться на тех товарах, которые ему действительно нужны и привлекательны для покупки. 
    Чтобы исправить ситуацию, можно на сайте настроить удаление неоплаченных товаров по истечении какого-то времени или стимуляции пользователя к покупке (прям так и писать, например, "мы видим, что товар Х у Вас в корзине уже больше 2-х недель". Если Вы еще его не куили, возможно, он попал в корзину по ошибке и Вам не очень нужен. Если это так, подтвердите, пожалуйста, его удаление из корзины, чтобы он не смущал ни Вас, ни нас, либо оплатите товар прямо сейчас [со скидкой ХХХ]". Кстати, доля покупок по акции - хоть и не такой важный, но тем не менее, тоже влияющий на классификацию целевого признак.
    4. маркет_актив_6_мес - еще один достаточно сильно влияющий на результат признак. Причем это именно активности за 6 месяцев, а не в текущем месяце. Возможно, это связано с инерционным мышлением и консерватизмом покупателей - им нужно время подумать, осознать, принять новую информацию о товарах, поэтому признак "количество маркетинговых коммуникаций в текущем месяце" не важен, а вот "количество маркетинговых коммуникаций за 6 мес" важно. Кроме того, возможно, имеет место накопительный эффект - чем больше ты будешь говорить, что "Тайд" - самый лучший порошок, тем выше вероятность, что рано или поздно люди в это поверят и купят его. Из этого следует вывод, что реклама все еще остается двигателем торговли, хотя и не самым главным в данном случае.
    5. Последний из наиболее важных признаков - средний_просмотр_категорий_за_визит (в теч. последнего месяца). Причем, чем больше категорий просматривает клиент, тем чаще признак склоняет модель классифицировать целевой к классу 0, т.е. активность НЕ снижается. Вывод - стимулировать просмотр разных категорий товаров для клиента, причем не важно каких, т.к. признаки категорий не влияют на оценку целевого признака.
